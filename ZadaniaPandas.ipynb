```python
chipotle = pd.read_csv('ML-datasets/chipotle.tsv', sep ='\t')
chipotle
```

```python
chipotle['price_dollars'] = chipotle['item_price']
chipotle
```

```python
dolarizer = lambda x: float(x[1:])
chipotle.price_dollars = chipotle.price_dollars.apply(dolarizer)
price_summed = chipotle.groupby('order_id').price_dollars.sum()
price_summed.max()
print(f'The highest price order is {price_summed.idxmax()} with ${price_summed.max()} ')
```
